```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('Barbara.bmp',0)
plt.imshow(img,cmap='gray')
```

```python
m=img.shape[0]
n=img.shape[1]
img_padded = np.zeros([m+2,n+2])
filter_padded = np.zeros([m+2,n+2])
```

```python
img_padded[:m,:n]=img
```

```python
filter_padded[511][511]=0
filter_padded[511][0]=1
filter_padded[511][1]=0
filter_padded[0][511]=1
filter_padded[0][0]=-4
filter_padded[0][1]=1
filter_padded[1][511]=0
filter_padded[1][0]=1
filter_padded[1][1]=0
```

```python
fft_filter = np.fft.fft2(filter_padded)
fft_img = np.fft.fft2(img_padded)
```

```python
plt.imshow(np.abs(fft_filter),cmap='gray')
```

```python
plt.imshow(1+np.log(np.abs(fft_img)), cmap='gray')
```

```python
plt.imshow(np.abs(np.fft.fftshift(fft_filter)),cmap='gray')
```

```python
plt.imshow(1+np.log(np.abs(np.fft.fftshift((fft_img)))), cmap='gray')
```

```python
res = fft_img * fft_filter
```

```python
ans = np.fft.ifft2(res)
```

```python
plt.imshow(np.real(ans),cmap='gray')
```
